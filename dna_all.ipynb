```python
import math
import pandas as pd

df_list = ['small', 'large']

for df in df_list:

    dataframe = pd.read_csv(f'./dna/databases/{df}.csv')

    if df == df_list[0]:
        txt_range = range(1,5)
    else:
        txt_range = range(5,21)

    for t in txt_range:
        print(f'DNA sample #{t}')

        with open (f'./dna/sequences/{t}.txt', 'r') as file:
            content = file.read()

        count = 0
        result = {}
        list_name = []

        for i in range(1,dataframe.shape[1]):
            for j in range(math.ceil(len(content)/4)):
                if (dataframe.columns[i]*j) in content:
                    temp = {dataframe.columns[i]: j}
                result.update(temp) 

        print(result)       

        for i in range(dataframe.shape[0]):
            print(dataframe.name[i])
            for j in range(1,dataframe.shape[1]):
                print(f'{dataframe.loc[i][dataframe.columns[j]]} vs. {result[dataframe.columns[j]]}')
                if dataframe.loc[i][dataframe.columns[j]] != result[dataframe.columns[j]]:
                    break
                else:
                    count += 1 
            if count == (dataframe.shape[1]-1):
                list_name = dataframe.name[i]
                break
            count = 0

        if not list_name:
            print(f'\nDNA sample #{t} does not have a match in the database.\n*****\n')
        else:
            print(f'\nDNA sample #{t} matches with {list_name}.\n*****\n')
```
